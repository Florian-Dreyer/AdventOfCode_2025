# Solution for Puzzle x

## Imports

```python
import pandas as pd
```

## Input Data

```python
input = pd.read_csv("input/day_5.csv", header=0)
print(f"Number of ranges: {len(input)}")
input.head(5)
```

```python
ranges = input.iloc[:182]
ids = input.iloc[182:]
```

## Part One

```python
number_valid_ids = 0
for id in ids["ids"].to_list():
    for r in ranges["ids"].to_list():
        start, end = r.split("-")
        start, end = int(start), int(end)
        if int(id) >= start and int(id) <= end:
            number_valid_ids += 1
            break

print(f"There are {number_valid_ids} valid ids.")
```

## Part Two

```python
ranges_combined = []
for r in ranges["ids"].to_list():
    start, end = r.split("-")
    start, end = int(start), int(end)
    ranges_combined.append((start, end))
ranges_combined.sort()
```

```python
idx = 0
while idx < len(ranges_combined)-1:
    if ranges_combined[idx][1] >= ranges_combined[idx+1][0]:
        ranges_combined[idx] = (ranges_combined[idx][0], max(ranges_combined[idx][1], ranges_combined[idx+1][1]))
        ranges_combined.pop(idx+1)
    else:
        idx += 1
```

```python
number_valid_ids = 0
for start, end in ranges_combined:
    number_valid_ids += (end-start+1)
print(f"There are {number_valid_ids} valid IDs")
```
