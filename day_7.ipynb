# Solution for Puzzle x

## Imports

```python
import pandas as pd
from collections import defaultdict
```

## Input Data

```python
input = pd.read_csv("input/day_7.csv", header=None)
print(f"Number of ranges: {len(input)}")
input.head(5)
```

## Part One

```python
beams = {input.iloc[0, 0].index('S')} # Position of S
number_of_splits = 0

# loop downwards
for idx, row in input[1:].iterrows():
    new_beams = set()
    for beam in beams:

        # check whether to split
        if row[0][beam] == "^":
            number_of_splits += 1
            if beam + 1 < len(row[0]):
                new_beams.add(beam + 1)
            if beam - 1 > -1:
                new_beams.add(beam - 1)
        else:
            new_beams.add(beam)
    beams = new_beams
print(f"There are {number_of_splits} total splits.")
```

## Part Two

```python
beams = {input.iloc[0, 0].index('S'): 1} # Position of S, now count number of possible origins for each stram

# loop downwards
for idx, row in input[1:].iterrows():
    new_beams = defaultdict(int)
    for beam in beams:

        # check whether to split
        if row[0][beam] == "^":
            beam_right = beam + 1
            if beam_right < len(row[0]):
                new_beams[beam_right] += beams[beam]
            beam_left = beam - 1
            if beam_left > -1:
                new_beams[beam_left] += beams[beam]
        else:
            new_beams[beam] += beams[beam]
    beams = new_beams
print(f"There are {sum(beams.values())} total timelines.")
```
