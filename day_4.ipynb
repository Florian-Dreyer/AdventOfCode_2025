# Solution for Puzzle 4

## Imports

```python
import pandas as pd
import itertools
```

## Input Data

```python
input = pd.read_csv("input/day_4.csv", header=0)
input = input[["rows"]]
print(f"Number of ranges: {len(input)}")
input.head(5)
```

## Part One

```python
# Transform input to 2d list
grid = []
for row in input["rows"].to_list():
    row_list = list(row)
    grid.append(row_list)
```

```python
accessible_rolls = 0
for i in range(len(grid)):
    for j in range(len(grid[0])):
        if grid[i][j] == "@":
            row_dirs = [-1, 0, 1]
            col_dirs = [-1, 0, 1]
            adjecent_rolls = 0

            # check all adjacent cells
            for dir in itertools.product(row_dirs, col_dirs):
                if dir == (0, 0):
                    continue
                x = i + dir[0]
                y = j + dir[1]

                # check if adjecent cell is in grid
                if x >= 0 and x < len(grid) and y >= 0 and y < len(grid[0]):
                    if grid[x][y] == "@":
                        adjecent_rolls += 1
            if adjecent_rolls < 4:
                accessible_rolls += 1
print(f"There are {accessible_rolls} accessible rolls.")
```
