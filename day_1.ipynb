# Solution for Puzzle 1

## Imports

```python
import pandas as pd
```

## Input data

```python
input = pd.read_csv("input/day_1.csv", header=0)
print(f"Number of rotations: {len(input)}")
input.head(5)
```

## Part One

```python
dial = 50
password = 0

# Iterator over all rotations
for rotation in input["rotations"]:
    direction, steps = rotation[0], int(rotation[1:])

    # Go all steps at once
    if direction == "R":
        dial += steps
    else:
        dial -= steps

    # Modulo replicates the dial logic
    dial = dial % 100
    if dial == 0:
        password += 1
print(f"The password is {password}")
```

## Part Two

```python
dial = 50
password = 0
history = {}
count = 0

# Iterator over all rotations
for rotation in input["rotations"]:
    direction = rotation[0]
    steps = int(rotation[1:])
    old_dial = dial
    old_pw = password

    # Adjust step direction
    if direction == "R":
        step = 1
    else:
        step = -1

    # Go through all steps
    for _ in range(steps):
        dial = (dial + step) % 100
        if dial == 0:
            password += 1
    history[count] = (password-old_pw, old_dial, step, steps)
    count += 1

print(f"The password is {password}")
```
