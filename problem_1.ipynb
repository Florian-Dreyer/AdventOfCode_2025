# Solution for Problem 1

## Imports

```python
import pandas as pd
```

## Input data

```python
input = pd.read_csv("input/problem_1.csv", header=0)
print(len(input))
input.head(5)
```

## Algorithm

```python
dial = 50
count = 0
for rotation in input["rotations"]:
    direction, steps = rotation[0], int(rotation[1:])
    if direction == "L":
        dial += steps
    else:
        dial -= steps
    dial = dial % 100
    if dial == 0:
        count += 1
print(f"The passwort is {count}")
```
