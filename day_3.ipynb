# Solution for Puzzle 3

## Imports

```python
import pandas as pd
```

## Input Data

```python
input = pd.read_csv("input/day_3.csv", header=0)
print(f"Number of banks: {len(input)}")
input.head(5)
```

## Part One

```python
maximum_outputs = []
for bank in input["banks"]:

    # largest element so far
    current_max_second = bank[-1]

    # current maximum output
    current_max_output = -float("inf")

    # go through bank in reverse order and memorize max element so far
    for joltage in bank[-2::-1]:
        curr = int(joltage+current_max_second)
        if curr > current_max_output:
            current_max_output = curr
        if int(joltage) > int(current_max_second):
            current_max_second = joltage
    maximum_outputs.append(current_max_output)
    
print(f"The maximum total output is {sum(maximum_outputs)}.")
```

## Part Two

```python
def compute_output(joltages: list) -> int:
    """Function to compute total output from twelve joltages."""
    if len(joltages) != 12:
        raise ValueError("Input joltages must have 12 elements!")
    return int("".join(joltages))

maximum_outputs = []
for bank in input["banks"]:
    # last 12 joltages as first try
    current_max_output = list(bank[-12:])

    # go through bank in reverse order
    for joltage in bank[-13::-1]:

        # if new joltage larger than previous replace it and let old element sink through
        curr = joltage
        curr_idx = 0
        while curr_idx < 12 and int(curr) >= int(current_max_output[curr_idx]):
            next_curr = current_max_output[curr_idx]
            current_max_output[curr_idx] = curr
            curr = next_curr
            curr_idx += 1
    maximum_outputs.append(compute_output(current_max_output))

print(f"The maximum total output is {sum(maximum_outputs)}.")
```
